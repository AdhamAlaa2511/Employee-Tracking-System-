```python
def clean_appname(name):
    name_lower = str(name).lower()
    for key in app_mapping.keys():
        if key in name_lower:
            return app_mapping[key][0]
    return "Unknown"

def categorized_app(name):
    for key in app_mapping.keys():
        if key.lower() in str(name).lower():
            return app_mapping[key][1]
    return 'Other'
```
